# Bitcoin Live Investment Simulator

The notebook scrapes the current Bitcoin-Price. Algorithms can be tested for investment strategies with this live-data.

```python
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
import random
import time

url = 'https://www.worldcoinindex.com/coin/bitcoin'

money = 10000
value = money
coins = 0
```

```python
wallet = []
for i in range(100):
    html = requests.get(url).text
    bs = BeautifulSoup(html, 'lxml')

    price = float(bs.find('td', {'class':'coinprice'}).get_text().strip().split('\n')[-1].replace(',', ''))
    
    if bool(random.getrandbits(1)):
        if coins==0:
            coins = money / price
            money = 0
        else:
            money = coins * price
            value = money
            coins = 0
            print('Money: {}, Price: {}'.format(money, price))
    wallet.append(value)
    time.sleep(1)
```

```python
plt.figure(figsize=(14,7))
plt.plot(wallet)
plt.title('Burning Money 2.0')
plt.xlabel('Transactions')
plt.ylabel('Bitcoins [€]')
plt.show()
```
